```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
A = (20, 35, 30, 35, 27)
B = (25, 32, 34, 20, 25)
ind = np.arange(N)

# Ustawienie kolorów dla poszczególnych kolumn
colors = ['pink', 'brown', 'purple', 'orange', 'red']
colors2 = ['purple', 'purple', 'purple', 'cyan', 'red']

# Rysowanie słupek dla chłopców
plt.bar(ind, A, label="A", color=colors[0:])

# Rysowanie słupek dla dziewczyn, z użyciem bottom=A, żeby umieścić je na poprzednich słupkach
plt.bar(ind, B, bottom=A, label="B", color=colors2[0:])

plt.ylabel('Wartości')
plt.title('Contribution by the teams')
plt.xticks(ind, ('Olsztyn', 'Kraków', 'Warszawa', 'Toruń', 'Gdańsk'))
plt.yticks(np.arange(0, 151, 10))
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Wczytaj dane z pliku Excel do ramki danych
df = pd.read_excel('/parki01.xlsx')

slask = df[df["Nazwa"]=="ŚLĄSKIE"]
podkarpackie = df[df["Nazwa"]=="PODKARPACKIE"]
lodzkie = df[df["Nazwa"]=="ŁÓDZKIE"]
x = slask["Rok"]
plt.plot(x, slask["Wartosc"], label='Śląsk', linestyle='-', linewidth=3)
plt.plot(x, podkarpackie["Wartosc"], label='Podkarpackike', linestyle='-', linewidth=3)
plt.plot(x, lodzkie["Wartosc"], label='lodzkie', linestyle='-', linewidth=3)
plt.xticks(x, slask["Rok"])
plt.text(max(x), max(slask["Wartosc"]), 'Aktualna data: 2030', fontsize=10, ha='right', va='top')
plt.tight_layout()
plt.grid()
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Wczytaj dane z pliku Excel do ramki danych
df = pd.read_csv('/medale01.csv', sep=';')
df["Rok"] = df["Rok"].astype(int)
df["Brązowe"] = df["Brązowe"].astype(int)
letnie = df[df["Rodzaj"] == "Letnie"]
zimowe = df[df["Rodzaj"] == "Zimowe"]
plt.subplot(1, 2, 1)
plt.pie(letnie["Brązowe"], labels=letnie["Miejsce"], autopct='%1.1f%%',shadow=True)
plt.title('Letnie igrzyska')

# Stwórz wykres kołowy dla zimowych igrzysk
plt.subplot(1, 2, 2)
plt.pie(zimowe["Brązowe"], labels=zimowe["Rok"], autopct='%1.1f%%',shadow=True)
plt.title('Zimowe igrzyska')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("sprzedaz03.xlsx")
data.columns = ["Produkt", "Rok", "Sprzedaż"]
r2015 = data[data["Rok"]== 2015]
r2017 = data[data["Rok"] == 2016]
r2018 = data[data["Rok"] == 2017]
x = np.arange(len(r2017["Produkt"]))
plt.barh(x, r2017["Sprzedaż"], label='2015', height=0.33)
plt.barh(x + 0.25, r2015["Sprzedaż"], label='2016', height=0.33)
plt.barh(x + 0.5, r2018["Sprzedaż"], label='2017', height=0.33)
plt.yticks(x + 0.33 / 2, r2017["Produkt"])
plt.legend()
plt.title('Dane samochodów w 2017 i 2018')
plt.tight_layout()
plt.show()
```
